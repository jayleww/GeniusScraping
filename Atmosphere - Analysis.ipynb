```python
#Import the packages to be used.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
#Import our data.
albumdf = pd.read_csv('atmosphere_album_data', header=0)
songdf = pd.read_csv('atmosphere_song_lyrics', header=0)
```

```python
#Remove unwanted columns. Need to address the index column being saved in the csv files.
albumdf.drop(['Unnamed: 0', 'Songs'], axis=1, inplace=True)
songdf.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
#Create a length feature for each song, as well as a count of unique words, remove x2 and 2x from the lyrics
from collections import Counter
remove_words = ['slug', 'hook', 'verse', 'x2', '2x']
for word in remove_words:
    songdf['SongLyrics'] = songdf['SongLyrics'].apply(lambda words: words.replace(word, ''))
#songdf['SongLyrics'] = songdf['SongLyrics'].apply(lambda words: words.replace('x2', ''))
#songdf['SongLyrics'] = songdf['SongLyrics'].apply(lambda words: words.replace('2x', ''))
songdf['SongLyricsNS'] = songdf['SongLyrics'].apply(lambda words: (' '.join([w for w in words.split() if w not in stopword_set])))
songdf['WordCount'] = songdf['SongLyrics'].map(len)
songdf['UniqueWords'] = songdf['SongLyrics'].apply(lambda words: len(list(Counter(words.split()).keys())))
```

```python
from nltk.corpus import stopwords
stopword_set = set(stopwords.words('english'))
#Create the same two columns, but with stopwords removed
songdf['WordCountNS'] = songdf['SongLyrics'].apply(lambda words: len(' '.join([w for w in words.split() if w not in stopword_set])))
songdf['UniqueWordsNS'] = songdf['SongLyrics'].apply(lambda words: len(list(Counter([w for w in words.split() if w not in stopword_set]).keys())))
```

```python
#Plot the new columns and compare them to the columns without stopwords
fig, ax = plt.subplots(ncols=3, sharey=True, figsize=(15,8))
songdf['WordCount'].hist(bins=20, alpha=0.7, ax=ax[0])
songdf['UniqueWords'].hist(bins=20, alpha=0.7, ax=ax[1])
ax[2].hist(x=100*(songdf['UniqueWords']/songdf['WordCount']), bins=20, alpha=0.7)
ax[0].set_xlabel('Number of Words in Song')
ax[1].set_xlabel('Number of Unique Words Used')
ax[2].set_xlabel('Pct of Unique Words')
songdf['WordCountNS'].hist(bins=20, color='red', alpha=0.5, ax=ax[0])
songdf['UniqueWordsNS'].hist(bins=20, color='red', alpha=0.5, ax=ax[1])
ax[2].hist(x=100*(songdf['UniqueWordsNS']/songdf['WordCountNS']), color='red', bins=20, alpha=0.5)
plt.suptitle('Word Counts')
plt.legend(('Stopwords', 'No Stopwords'))
```

So by removing the stopwords we reduce the number of words and number of unique words used, but the percentage
of unique words rises.

```python
#Lets make that percentage of unique words a new feature column
songdf['PctUnique'] = round(100*(songdf['UniqueWordsNS']/songdf['WordCountNS']), 3)
```

```python
#Now I want to create a new df with information from the albumdf joined based on AlbumID.
joindf = songdf.merge(albumdf, on='AlbumID')
joindf.to_csv('atmospheredf', header=True)
joindf
```

```python
############################### Sentiment Analysis ####################################
#Download the lexicon to be used
nltk.download('vader_lexicon')
```

```python
#Import sentiment package
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
#Check that the sentiment analyzer is working correctly
nltk_sentiment = SentimentIntensityAnalyzer()
score = nltk_sentiment.polarity_scores(joindf['SongLyricsNS'][0])
score
```

```python
#Create a dataframe with the song names/lyrics and the scores
nltk_results = [nltk_sentiment.polarity_scores(row) for row in joindf['SongLyrics']]
sentiment_df = pd.DataFrame(nltk_results)
sentiment_df = pd.concat([joindf['SongName'], sentiment_df], axis=1)
resultsdf = pd.merge(joindf, sentiment_df, on='SongName')
resultsdf.head()
```

```python
#Find the top 5 sentiments based on the compound scores
resultsdf.sort_values(by='compound', ascending=False).head(5)
```

```python
#Find the lowest 5 sentiments based on the compound scores
resultsdf.sort_values(by='compound', ascending=True).head(5)
```

```python
#Take a look at the distributions from the vader sentiment analysis
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(16,8))
resultsdf['compound'].hist(bins=40, ax=ax[0,0])
ax[0,0].set_xlabel('Compound')
resultsdf['neg'].hist(bins=40, color='red', ax=ax[0,1])
ax[0,1].set_xlabel('Negative')
resultsdf['neu'].hist(bins=40, color='grey', ax=ax[1,0])
ax[1,0].set_xlabel('Neutral')
resultsdf['pos'].hist(bins=40, color='green', ax=ax[1,1])
ax[1,1].set_xlabel('Positive');
fig.suptitle('Distributions of the Sentiment Analysis');
```

```python
#Heatmap
plt.figure(figsize=(12,6))
sns.heatmap(resultsdf[['compound', 'neg', 'pos', 'neu']], )
```

```python
#Compound column looks extremely polarizing. Going to try and cluster based only on the neg, pos, neu columns.
sent_X = resultsdf[['neg', 'pos', 'neu']]
```

```python
#Check optimal cluster number using the elbow method
wcss = []
for i in np.arange(1,15,1):
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, verbose=0)
    km.fit(sent_X)
    wcss.append(km.inertia_)
```

```python
#Plot the elbow method
plt.scatter(x=np.arange(1,15,1), y=wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS');
```

```python
#Fit KMeans using 6 clusters.
km = KMeans(n_clusters=6, init='k-means++', n_init=10, verbose=0)
km.fit(sent_X)
```

```python
resultsdf['Cluster'] = km.predict(sent_X)
```

```python
#Looks almost all the clusters do actually catch many different songs.
resultsdf['Cluster'].value_counts()
```

```python
#Look for a correlation between the cluster numbers and the neg/pos/neu columns.
corr = resultsdf[['Cluster', 'neg', 'pos', 'neu', 'compound']].corr()
sns.heatmap(corr, cbar=False, annot=True)
```

```python
########Topic modelling#######

#Start by vectorizing the corpus
from sklearn.feature_extraction.text import CountVectorizer
#Bag-of-words from all the lyrics
word_matrix = CountVectorizer()
words = word_matrix.fit_transform(resultsdf['SongLyricsNS'])
```

```python
#Create the model and fit it to the matrix
from sklearn.decomposition import LatentDirichletAllocation
#Use 6 topics to match the clustering
lda = LatentDirichletAllocation(n_components=6)
#Fit the model to the word matrix
lda.fit(words);
```

```python
#Print off the 6 topics
feature_names = word_matrix.get_feature_names()
for topic_idx, topic in enumerate(lda.components_):
    message = "Topic #%d: " % topic_idx
    message += " ".join([feature_names[i] for i in topic.argsort()[:-20 - 1:-1]])
    print(message)
```
